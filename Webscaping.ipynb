```python
import requests
from bs4 import BeautifulSoup

url = "https://www.spam.com"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup.find_all("a")
link_list =[l.get('href') for l in soup.findAll("a")]
print(link_list)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.simplyrecipes.com/recipes/chicken_tikka_masala/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
pSoup = soup.find_all("li")
for div in pSoup:
    print(div.text)
	#returns true if tag has attribute called sample
    if div.has_attr("_class"):
          if div["class"] == "structured-ingredients__list-item":
                print(div.text)
```
